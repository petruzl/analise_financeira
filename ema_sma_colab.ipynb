```python
import yfinance as yf
import ta
import pandas as pd
import plotly.graph_objects as go
```

```python
ativo = '^BVSP'
dados = yf.download(ativo, start='2022-01-01', end='2023-08-01')
dados
```

```python
# MÉDIA MÓVEL SIMPLES - SMA
periodo_sma = 20
dados['Sma'] = ta.trend.sma_indicator(dados['Close'], window=periodo_sma)
dados
dados.head(30)
fig = go.Figure()
df_fig = dados
fig.add_trace(go.Candlestick(
    x=df_fig.index,
    open=df_fig['Open'],
    high=df_fig['High'],
    low=df_fig['Low'],
    close=df_fig['Close'],
    name='Grafico de Candlesticks com SMA'))
fig.add_trace(go.Scatter(name='Sma',
                         x=df_fig.index,
                         y=dados['Sma'],
                         marker_color='black'))
fig.update_layout(template='ggplot2', width=1400, height=700)
fig.show()
```

```python
# MEDIA MÓVEL EXPONENCIAL
periodo_ema = 50
dados['Ema'] = ta.trend.ema_indicator(dados['Close'], window=periodo_ema)
fig = go.Figure()
df_fig = dados
fig.add_trace(go.Candlestick(
                              x=df_fig.index,
                              open=df_fig['Open'],
                              high=df_fig['High'],
                              low=df_fig['Low'],
                              close=df_fig['Close'],
                              name='Grafico de Candlesticks com EMA'))
fig.add_trace(go.Scatter(name='Sma',
                         x=df_fig.index,
                         y=dados['Sma'],
                         marker_color='black'))
fig.add_trace(go.Scatter(name='Ema',
                         x=df_fig.index,
                         y=dados['Ema'],
                         marker_color='blue'))
fig.update_layout(template='ggplot2', width=1400, height=700)
fig.show()
```
