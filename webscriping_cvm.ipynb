```python
import pandas as pd
import numpy as np
import requests
```

```python
link = 'https://dados.cvm.gov.br/dados/CIA_ABERTA/CAD/DADOS/cad_cia_aberta.csv'
```

```python
r = requests.get(link)
```

```python
linhas = []

for i in r.text.split('\n'):
   linhas.append(i.strip().split(';'))
```

```python
data = pd.DataFrame(linhas[1:], columns = linhas[0])
```

```python
data.head()
```

```python
#list comprehension
lines = [i.strip().split(';') for i in r.text.split('\n')]
```

```python
data = pd.DataFrame(lines[1:], columns = lines[0] )
```

```python
data.head()
```

```python
serchfor = ['MULTIPLAN', 'MALL', 'IGUATEMI', 'SHOPP']
shoppings = data[data.DENOM_SOCIAL.str.contains('|'.join(serchfor), na = False)]
```

```python
shoppings
```

```python
shoppings_ativos = shoppings[shoppings['SIT'] == 'ATIVO']
```
